```python
# WAP to print sum of digit of a given number
#input = 123 output = 6
```

```python
n=int(input('Enter a number: '))
sum=0
while n!=0:
    d=n%10
    sum =sum+d
    n=n//10
print('Sum',sum)    
```

```python
#WAP to reverse a number
# input-123 output-321
```

```python
n=int(input('Enter a number: '))
reverse=0
while n!=0:
    d=n%10
    reverse=reverse*10+d
    n=n//10
print('Reverse',reverse)
```
