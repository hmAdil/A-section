```python
def readfile():
    with open('music_festival.txt','r') as f:
        rows=f.readlines()
    return rows
def subprob1():
    rows=readfile()
    genres=[]
    count=0
    for line in rows:
        count +=1
        rec=line.split(',')
        print(rec)
        genre=rec[1].strip() #remove unnecessary spacing
        print(genre)
        if genre not in genres:
            genres.append(genre)
    return(count,len(genres))

def subprob2(event_type,arist,genre,date,time,stage):
    new_event=f"{event_type}:{artist},{genre},{date},{time},{stage}\n"
    rows=readfile()
    with open('music_festival.txt','a') as f1:
        if not rows[-1].endswith('\n'):
            f1.write('\n')
        f1.write(new_event)
    print('New event inserted.')

def subprob3():
    rows=readfile()
    genre_count ={}
    for line in rows:
        genr=line.split(',')
        #print(genr)
        genre=genr[1].strip() #remove unnecessary spacing
        if genre in genre_count:
            genre_count[genre]+=1
        else:
            genre_count[genre]=1
    if genre_count:
        print("Event count by genre")
        for genre,count in genre_count.items():
            print(f'{genre}:{count}')
    else:
        print('No data to display:')

def subprob4(artist_name):
    rows=readfile()
    flag=0
    for line in rows:
        artist=line.split(',')[0].split(':')[1].strip()
        if artist.lower()==artist_name.lower():
            rows.remove(line)
            flag=1
        with open ('music_festival.txt','w') as f1:
            f1.writelines(rows)

    if flag:
        print(f'event by {artist_name} is removed.')
    else:
        print(f"Event by {artist_name} wasn't found")



        
def subprob5():
    time_count={}
    rows=readfile()
    for line in rows:
        time=line.split(',')[3].strip()
        if time in time_count:
            time_count[time]+=1
        else:
            time_count[time]=1

    if time_count:
        for time, count in time_count.items():
            print(f'{time}:{count} events')
    else:
        print('No Events Found.')

def subprob6():
    rows=readfile()
    genre_summary={}
    for line in rows:
        rec=line.split(',')
        genre=rec[1].strip()
        event_type=rec[0].split(':')[0].strip()
        artist=rec[0].split(':')[1].strip()
        if genre not in genre_summary:
            genre_summary[genre]=[]
            genre_summary[genre].append(f'{event_type}by{artist}')
    if genre_summary:
        print('total events:',sum(len(events) for events in genre_summary.values()))
    print(genre_summary)

def subprob7():
    open("music_festival.txt",'w').close()
    print("All events are deleted.")
    

def return_artist(x):
    return x[0].split(':')[1].strip()


def subprob8():
    rows=readfile()
    events=[]
    for line in rows:
        rec=line.split(',')
        events.append(rec)
    if events:
        events.sort(key=return_artist)
        for event in events:
            print(','.join(event).strip())
    else:
        print('No Events to display.')
        

while True:
    print('''Hello!
    Possible Choices for Subproblem: 
    0 to exit 
    1 to 8''')
    ch=int(input("Enter Choice:"))
    if ch==1:
        total_events,unique_genres=subprob1()
        if total_events==0 and unique_genres==0:
            print('No Data!')
        else:
            print(f'Total events are {total_events}.')
            print(f'Unique genres are {unique_genres}.\n')
    elif ch==2:
        event_type=input('Enter event type: ').strip()
        artist=input('Enter artist: ').strip()
        genre=input('Enter genre:').strip()
        date=input('Enter date as YYYY-MM-DD:').strip()
        time=input('Enter time as HH:MM AM or PM:').strip()
        stage=int(input('Enter the stage number:'))
        subprob2(event_type,artist,genre,date,time,stage)
    elif ch==0:
        print("Goodbye!")
        break
    elif ch==3:
        subprob3()
        print('\n')
    elif ch==4:
        artist_name=input('Enter artist name:').strip()
        subprob4(artist_name)
        print('\n')
    elif ch==5:
        subprob5()
        print('\n')
    elif ch==6:
        subprob6()
        print('\n')
    elif ch==8:
        subprob8()
        print('\n')
    elif ch==7:
        subprob7()
        print('\n')
    else:
        print('''Invalid choice!
        Choose between 0 or 1.\n''')
```
